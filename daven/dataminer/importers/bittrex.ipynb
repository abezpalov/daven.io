```python
# Функция загрузки открытых данных
base_url = 'https://bittrex.com/api/v1.1'

def load(method, pair = None, limit = None, timeout = 10):

    if pair and limit:
        url = '{}/{}/?market={}&limit={}'.format(base_url, method, pair, limit)
    elif pair:
        url = '{}/{}?market={}'.format(base_url, method, pair)
        print(url)
    else:
        url = '{}/{}'.format(base_url, method)

    r = requests.get(url, timeout = timeout)

    result = r.json()
    if result['success'] == True:
        return result['result']
    else:
        return None
```

```python
method = 'public/getmarkets'
pairs_ = load(method)

pairs_
```

```python
for pair_ in pairs_:
    print(pair_['MarketName'])
```

```python
method = 'public/getcurrencies'
currencies_ = load(method)

currencies_
```

```python
for currency_ in currencies_:
    print(currency_['Currency'], currency_['CurrencyLong'])
```

```python
method = 'public/getticker'

for pair_ in pairs_:
    ticker_ = load(method, pair_['MarketName'])
    print(ticker_)
```
